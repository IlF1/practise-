```

```



````python
a = [1,2,3,4,7]
b = [3,4,5,6,9]
def elements(a,b):
  aa1=set(a)
  bb1=set(b)
  unik1=aa1-bb1
  unik2=bb1-aa1
  allunik=unik1.union(unik2)
  return list(allunik)
print(elements(a,b))





````

````python
# import random


# def createknight():
#   name=input('введи имя персонажа ')
#   health=input("введи инфу про здоровье(MAX=1000 health points) ")
#   armour=int(input("введи количество едениц брони(MAX=300)" ))
#   minhit=int(input("введи минимальный урон (min =100)" ))
#   maxhit=int(input("введи максимальный урон(max=250)" ))
#   return {
#        "имя" : name,
#        "хп": health,
#        "броня": armour,
#        "урон":(minhit,maxhit)
#        }




# def createdragon():
#   name=input('введи имя драконаа ')
#   health=input("введи инфу про здоровье(MAX=5000 health points) ")
#   minhit=int(input("введи минимальный урон (min =30)" ))
#   maxhit=int(input("введи максимальный урон(max=80)" ))
#   return {
#        "имя": name,
#        "хп" :health,
#        "урон":(minhit,maxhit)
#        }




# def attack(character):
#     damage = random.randint(*character['урон'])
#     print(f"{character['имя']} атакует и наносит {damage} урон")
#     return damage



# def take_damage(character, damage):
#     effective_damage = max(0, damage - character['броня'] if 'броня' in character else damage)
#     character['хп'] -= effective_damage
#     print(f"{character['имя']} получает {effective_damage} урона! Осталось здоровья: {character['хп']}")


# def is_alive(character):
#     return character['хп'] > 0


# def battle(knight, dragon):
#     print("начало боя")
#     while is_alive(knight) and is_alive(dragon):
#         action = input("Выберите действие: 1- Атаковать")

#         if action == "1":
#             damage_to_dragon = attack(knight)
#             take_damage(dragon, damage_to_dragon)

#             if is_alive(dragon):
#                 damage_to_knight = random.randint(*dragon['урон'])
#                 take_damage(knight, damage_to_knight)
#         # elif action == "2":
#         #     print("Вы убегаете от дракона!")
#         #     break
#         else:
#             print("Неверное действие. Попробуйте снова.")

#     if  not is_alive(knight):
#         print("Рыцарь пал в бою! Игра окончена.")
#     elif not is_alive(dragon):
#         print("Дракон повержен! Вы победили!")




# def main():
#   print("Вы в игре Рыцарь и Дракон")
#   knight=createknight()
#   dragon=createdragon()

#   battle(knight, dragon)

# main()



import random


def create_knight():
    name = input('Введите имя персонажа: ')
    health = int(input("Введите здоровье (MAX=1000 health points): "))
    while health > 1000 or health < 0:
        health = int(input("Здоровье должно быть в диапазоне от 0 до 1000. Попробуйте снова: "))

    armour = int(input("Введите количество единиц брони (MAX=300): "))
    while armour > 300 or armour < 0:
        armour = int(input("Броня должна быть в диапазоне от 0 до 300. Попробуйте снова: "))

    minhit = int(input("Введите минимальный урон (min =100): "))
    maxhit = int(input("Введите максимальный урон (max=250): "))
    while minhit < 100 or maxhit > 250 or minhit > maxhit:
        minhit = int(input("Минимальный урон должен быть не менее 100, максимальный не более 250 и меньше минимального. Попробуйте снова: "))
        maxhit = int(input("Введите максимальный урон (max=250): "))

    return {
        "имя": name,
        "хп": health,
        "броня": armour,
        "урон": (minhit, maxhit)
    }


def create_dragon():
    name = input('Введите имя дракона: ')
    health = int(input("Введите здоровье (MAX=5000 health points): "))
    while health > 5000 or health < 0:
        health = int(input("Здоровье должно быть в диапазоне от 0 до 5000. Попробуйте снова: "))

    minhit = int(input("Введите минимальный урон (min =30): "))
    maxhit = int(input("Введите максимальный урон (max=80): "))
    while minhit < 30 or maxhit > 80 or minhit > maxhit:
        minhit = int(input("Минимальный урон должен быть не менее 30, максимальный не более 80 и меньше минимального. Попробуйте снова: "))
        maxhit = int(input("Введите максимальный урон (max=80): "))

    return {
        "имя": name,
        "хп": health,
        "урон": (minhit, maxhit)
    }


def attack(character):
    damage = random.randint(*character['урон'])
    print(f"{character['имя']} атакует и наносит {damage} урона.")
    return damage


def take_damage(character, damage):
    effective_damage = max(0, damage - character.get('броня', 0))
    character['хп'] -= effective_damage
    character['хп'] = max(0, character['хп'])  # Убедитесь, что здоровье не становится отрицательным
    print(f"{character['имя']} получает {effective_damage} урона! Осталось здоровья: {character['хп']}")


def is_alive(character):
    return character['хп'] > 0


def battle(knight, dragon):
    print("Начало боя!")
    while is_alive(knight) and is_alive(dragon):
        action = input("Выберите действие: 1 - Атаковать: ")

        if action == "1":
            damage_to_dragon = attack(knight)
            take_damage(dragon, damage_to_dragon)

            if is_alive(dragon):
                damage_to_knight = attack(dragon)
                take_damage(knight, damage_to_knight)
        else:
            print("Неверное действие. Попробуйте снова.")

    if not is_alive(knight):
        print("Рыцарь пал в бою! Игра окончена.")
    elif not is_alive(dragon):
        print("Дракон повержен! Вы победили!")


def main():
    print("Вы в игре Рыцарь и Дракон")
    knight = create_knight()
    dragon = create_dragon()
    battle(knight, dragon)


main()

````

# Задание 3


Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests


def get_pokemon_data(name):
    url = f"https://pokeapi.co/api/v2/pokemon/{name}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Ошибка: Покемон '{name}' не найден.")
        return None


def add_pokemon(team, name):
    if any(p['name'] == name for p in team):
        print(f"Покемон '{name}' уже есть в команде.")
        return

    data = get_pokemon_data(name)
    if data:
        team.append({
            "name": data['name'],
            "height": data['height'],
            "weight": data['weight'],
            "types": [t['type']['name'] for t in data['types']],
            "stats": {stat['stat']['name']: stat['base_stat'] for stat in data['stats']}
        })
        print(f"Покемон '{name}' добавлен в команду.")
    else:
        print(f"Не удалось добавить покемона '{name}'.")


def remove_pokemon(team, name):
    for pokemon in team:
        if pokemon['name'] == name:
            team.remove(pokemon)
            print(f"Покемон '{name}' удален из команды.")
            return
    print(f"Покемон '{name}' не найден в команде.")


def view_team(team):
    if not team:
        print("Ваша команда пуста.")
        return
    print("Ваша команда покемонов:")
    for pokemon in team:
        print(f"{pokemon['name']} - Высота: {pokemon['height']}, Вес: {pokemon['weight']}, Типы: {', '.join(pokemon['types'])}")


def find_pokemon(team, name):
    for pokemon in team:
        if pokemon['name'] == name:
            print(f"Найден покемон: {pokemon['name']} - Высота: {pokemon['height']}, Вес: {pokemon['weight']}, Типы: {', '.join(pokemon['types'])}")
            return
    print(f"Покемон '{name}' не найден в команде.")


def battle(team, name1, name2):
    if not team:
        print("Ваша команда пуста. Невозможно устроить бой.")
        return

    pokemon1 = next((p for p in team if p['name'].lower() == name1.lower()), None)
    pokemon2 = next((p for p in team if p['name'].lower() == name2.lower()), None)

    if not pokemon1 or not pokemon2:
        print("Один или оба покемона не найдены в команде.")
        return

    stats1 = pokemon1["stats"]
    stats2 = pokemon2["stats"]

    print(f"Бой между {pokemon1['name']} и {pokemon2['name']}!")

    attack1 = stats1.get('attack', 0)
    attack2 = stats2.get('attack', 0)

    if attack1 > attack2:
        print(f"{pokemon1['name']} победил!")
    elif attack1 < attack2:
        print(f"{pokemon2['name']} победил!")
    else:
        print("Ничья!")


def main():
    team = []
    while True:
        print("Менеджер команды Pokémon")
        print("1. Добавить покемона")
        print("2. Удалить покемона")
        print("3. Просмотреть команду")
        print("4. Найти покемона по имени")
        print("5. Устроить бой между покемонами")
        print("6. Выход")

        choice = input("Выберите действие: ")

        if choice == "1":
            name = input("Введите имя покемона для добавления: ")
            add_pokemon(team, name)
        elif choice == "2":
            name = input("Введите имя покемона для удаления: ")
            remove_pokemon(team, name)
        elif choice == "3":
            view_team(team)
        elif choice == "4":
            name = input("Введите имя покемона для поиска: ")
            find_pokemon(team, name)
        elif choice == "5":
            name1 = input("Введите имя первого покемона: ")
            name2 = input("Введите имя второго покемона: ")
            battle(team, name1, name2)
        elif choice == "6":
            print("Выход из программы.")
            break
        else:
            print("Неверный выбор, попробуйте снова.")


main()


# import requests


# def pokemons_requests(name):
#   url = f"https://pokeapi.co/api/v2/pokemon/{name}"
#   response=request.get(url)
#   status_code=response.status_code
#   if status_code==200:
#     return response.json()
#   else:
#     print(f"Ошибка: Покемон '{name}' не найден.")
#     print(status_code)


# def add_pokemon(team,name):
#   for pokemon in team:
#       if name not in pokemon['name']:
#         data = pokemons_requests(name)
#         if data:
#             team.append({
#                 "name": data['name'],
#                 "height": data['height'],
#                 "weight": data['weight'],
#                 "types": [t['type']['name'] for t in data['types']],
#                 "stats": {stat['stat']['name']: stat['base_stat'] for stat in data['stats']}
#             })
#             print(f"Покемон '{name}' добавлен в команду.")
#         else:
#             print(f"Не удалось добавить покемона '{name}'.")
#       else:
#         print(f"Покемон '{name}' уже есть в команде.")


# def remove_pokemon(team, name):
#     for pokemon in team:
#         if pokemon['name'] == name:
#             team.remove(pokemon)
#             print(f"Покемон '{name}' удален из команды.")
#             return
#     print(f"Покемон '{name}' не найден в команде.")


# def view_team(team):
#     if not team:
#         print("Ваша команда пуста.")
#     else:
#       print("Ваша команда покемонов:")
#       for pokemon in team:
#         print(f"{pokemon['name']} - Высота: {pokemon['height']}, Вес: {pokemon['weight']}, Типы: {', '.join(pokemon['types'])}")






# def main():
#     team = []
#     while True:
#         print("Менеджер команды Pokémon")
#         print("1. Добавить покемона")
#         print("2. Удалить покемона")
#         print("3. Просмотреть команду")
#         print("4. Найти покемона по имени")
#         print("5. Устроить бой между покемонами")
#         print("6. Выход")

#         choice = input("Выберите действие: ")

#         if choice == "1":
#             name = input("Введите имя покемона для добавления: ")
#             add_pokemon(team, name)
#         elif choice == "2":
#             name = input("Введите имя покемона для удаления: ")
#             remove_pokemon(team, name)
#         elif choice == "3":
#             view_team(team)
#         elif choice == "4":
#             name = input("Введите имя покемона для поиска: ")
#             find_pokemon(team, name)
#         elif choice == "5":
#             name1 = input("Введите имя первого покемона: ")
#             name2 = input("Введите имя второго покемона: ")
#             battle(team, name1, name2)
#         elif choice == "6":
#             print("Выход из программы.")
#             break
#         else:
#             print("Неверный выбор, попробуйте снова.")


# main()


````
