```python
# import requests
# url = "https://api.artic.edu/api/v1/artworks"
# page = int(input("Введи страницу с изображениями(MAX 100)" ))
# limit=int(input("Введи лимит изображений на страницу(MAX 99)"))


# def art_list(url, page, limit): #Извлекает список произведений искусства из API.
#   response = requests.get(f"{url}?page={page}&limit={limit}")
#   if response.status_code == 200:
#         data = response.json()
#         artworks = data['data']
#         pagination = data['pagination']
#         return artworks, pagination
#   else:
#         print("Ошибка при извлечении данных:", response.status_code)
# print(art_list(url, page, limit))


import requests

url= "https://api.artic.edu/api/v1/artworks"
#Извлекает список произведений искусства из API.
def fetch_artworks(page=1):
    response = requests.get(f"{url}?page={page}&limit=15")
    data = response.json()
    return data['data'], data['pagination']

#Фильтрует список произведений искусства по имени художника.
def filter_artworks_by_artist(artworks, artist_name):
    filtered_artworks = [
        artwork for artwork in artworks
        if 'artist_display' in artwork and artist_name.lower() in artwork['artist_display'].lower()
    ]
    return filtered_artworks
#"Отображает подробную информацию о произведении искусства.
def display_artwork_details(artwork):
    title = artwork.get('title', 'Без названия')
    artist = artwork.get('artist_display', 'Неизвестен')
    date = artwork.get('date_display', 'Неизвестна дата')
    medium = artwork.get('medium_display', 'Неизвестен носитель')

    print(f"nНазвание: {title}")
    print(f"Исполнитель: {artist}")
    print(f"Дата: {date}")
    print(f"Носитель: {medium}")

def main():
    current_page = 1
    while True:
        artworks, pagination = fetch_artworks(current_page)

        print(f"nСтраница {current_page} из {pagination['total_pages']}")
        for index, artwork in enumerate(artworks):
            print(f"{index + 1}. {artwork['title']} - {artwork.get('artist_display', 'Неизвестен')}")

        print("nОпции:")
        print("1. Выбрать произведение искусства для просмотра деталей")
        print("2. Перейти на следующую страницу")
        print("3. Перейти на предыдущую страницу")
        print("4. Фильтровать по исполнителю")
        print("5. Выход")
        choice = input("Выберите опцию: ")
        if choice == '1':
            artwork_index = int(input("Введите номер произведения: ")) - 1
            if 0 <= artwork_index < len(artworks):
                display_artwork_details(artworks[artwork_index])
            else:
                print("Неверный номер произведения.")

        elif choice == '2':
            if current_page < pagination['total_pages']:
                current_page += 1
            else:
                print("Это последняя страница.")

        elif choice == '3':
            if current_page > 1:
                current_page -= 1
            else:
                print("Это первая страница.")

        elif choice == '4':
            artist_name = input("Введите имя художника: ")
            filtered_artworks = filter_artworks_by_artist(artworks, artist_name)
            if filtered_artworks:
                for index, artwork in enumerate(filtered_artworks):
                    print(f"{index + 1}. {artwork['title']} - {artwork.get('artist_display', 'Неизвестен')}")
            else:
                print("Произведения искусства не найдены для данного художника.")

        elif choice == '5':
            print("Выход из программы.")
            break

        else:
            print("Неверный выбор. Пожалуйста, попробуйте снова.")
print(main())






```

# Задание 3

Задача: Создать программу по управлению портфелем криптовалют

Цель: Создать скрипт на Python, который извлекает цены на криптовалюты в режиме реального времени, позволяет пользователям управлять портфелем криптовалют, вычисляет общую стоимость портфеля, отслеживает изменения цен и предоставляет исторические данные о ценах для анализа.

Требования:
Получение текущих цен на криптовалюты:

Используйте https://docs.coingecko.com/ для получения актуальных цен на список криптовалют.

Управление портфелем:

- Позволяет пользователю создавать портфель криптовалют и управлять им, указывая количество каждой криптовалюты, которой он владеет.
- Расчитывает общую стоимость портфеля в указанной фиатной валюте (например, долларах США).

Отслеживание изменения цен:

- Отображение процентного изменения цены для каждой криптовалюты в портфеле за последние 24 часа.
- Выделите все криптовалюты, стоимость которых значительно увеличилась или снизилась.

Поиск исторических данных о ценах:

- Получение исторических данных о ценах на указанную криптовалюту за последнюю неделю.
- Предоставьте пользователю возможность визуализировать эти данные в простом текстовом формате (например, цены за день).

Взаимодействие с пользователем:

- Реализуйте интерфейс командной строки для ввода данных пользователем.
- Предоставьте опции для получения текущих цен, управления портфелем, просмотра изменений цен или анализа исторических данных.


```python
import requests
from datetime import datetime, timedelta
url="https://api.coingecko.com/api/v3"
portfolio={}

#Добавить криптовалюту в портфель
def crypto_case(name,kol):
    if name in portfolio:
        portfolio[name] += kol
    else:
        portfolio[name] = kol
    print(f"Добавлено {kol} {name} в портфель.")


#Удалить криптовалюту из портфеля
def remove_crypto(name, kol):
    if name in portfolio:
        portfolio[name] = portfolio[name] - kol
        if portfolio[name] <= 0:
            del portfolio[name]
        print(f"Удалено {kol} {name} из портфеля.")
    else:
        print(f"{name} не найден в портфеле.")


#Получить текущую цену криптовалюты
def get_current_price(name, fiat_currency='usd'):
    response = requests.get(f"{url}/simple/price?ids={name}&vs_currencies={fiat_currency}&include_24hr_change=true")
    status_code = response.status_code
    if status_code == 200:
        data = response.json()
        price = data[name][fiat_currency]
        change_percentage = data[name][f'{fiat_currency}_24h_change']
        return price, change_percentage
    else:
        print(f"Ошибка при получении данных для {name}. Статус код: {status_code}")
        return None, None


#Получить общую стоимость портфеля
def get_portfolio_value(fiat_currency='usd'):
    total_value = 0
    for name, kol in portfolio.items():
        price, change_percentage = get_current_price(name, fiat_currency)
        if price is None:  # Проверяем на None
            print(f"Не удалось получить цену для {name}.")
            continue  # Пропускаем эту криптовалюту
        total_value += price * kol
    return total_value

#Получить исторические данные о ценах криптовалюты за последнюю неделю.
def get_historical_data(name, fiat_currency='usd'):
    end_date = int(datetime.now().timestamp())
    start_date = int((datetime.now() - timedelta(days=7)).timestamp())
    response = requests.get(f"{url}/coins/{name}/market_chart/range?vs_currency={fiat_currency}&from={start_date}&to={end_date}")
    status_code=response.status_code
    if status_code==200:
        data = response.json()
        prices = data['prices']
        historical_prices = {datetime.fromtimestamp(price[0] / 1000).strftime('%Y-%m-%d'): price[1] for price in prices}
        return historical_prices
    else:
        return("Ошибка", status_code)


def main():
  while True:
        print("доступные варианты:")
        print("1. Добавить криптовалюту в портфель")
        print("2. Удалить криптовалюту из портфеля")
        print("3. Показать общую стоимость портфеля")
        print("4. Показать изменения цен за 24 часа")
        print("5. Получить исторические данные о ценах")
        print("6. Выход")
        choice = input("Выберите опцию: ")

        if choice == '1':
            name = input("Введите название криптовалюты (например, bitcoin): ")
            kol = float(input("Введите количество: "))
            crypto_case(name, kol)
        elif choice == '2':
            symbol = input("Введите символ криптовалюты для удаления: ")
            amount = float(input("Введите количество для удаления: "))
            remove_crypto(symbol, amount)
        elif choice == '3':
            fiat_currency = input("Введите фиатную валюту (например, usd): ")
            total_value = get_portfolio_value(fiat_currency)
            print(f"Общая стоимость портфеля: {total_value:.2f} {fiat_currency}.")
        elif choice == '4':
            for name in portfolio.keys():
                current_price, change_percentage = get_current_price(name)
                print(f"{name}: Текущая цена: {current_price}, Изменение за 24 часа: {change_percentage:.2f}%")
        elif choice == '5':
            name= input("Введите название криптовалюты для получения исторических данных: ")
            historical_data = get_historical_data(name)
            print(f"исторические данные для {name}:")
            for date, price in historical_data.items():
                print(f"{date}: {price:.2f} USD")
        elif choice == '6':
            print("Выход из программы.")
            break
print(main())

```
